### 4.4: Extract the time series data

If the response status is OK (200), we need to extract the time series data from the response.

````python
# Loop over stock in stocks to retrieve data from alpha vantage using my api key
for stock in stocks:
    api_url = f'{base_url}function={function}&symbol={stock}&apikey={api_key}' 
    response = requests.get(api_url)
    if response.status_code == 200:
        data = response.json().get('Time Series (Daily)', {})
        for date, values in data.items():
            time_series_data.append({
                "date": date,
                "ticker": stock,
                "1. open": values["1. open"],
                "2. high": values["2. high"],
                "3. low": values["3. low"],
                "4. close": values["4. close"],
                "5. volume": values["5. volume"]
            })
    else:
        print(f"Failed to retrieve data for {stock}.")

# Convert the list into a pandas DataFrame
df = pd.DataFrame(time_series_data)
````

### Final DF

````python
# Remove numbers and periods from column names and convert to lowercase
df.columns = df.columns.str.replace('[0-9.]', '', regex=True).str.strip().str.lower()

# Convert 'date' column to datetime type
df['date'] = pd.to_datetime(df['date'])

df
````

  async function fetchStockData(symbol) {
    try {
      const jsonFileName = symbol.toLowerCase() + ".json";
      const res = await fetch(jsonFileName);
      const data = await res.json();
      const timeSeries = data["Time Series (Daily)"];
      const sortedDates = Object.keys(timeSeries).sort();
      const dates = sortedDates.slice(-30);
      const prices = dates.map(date => parseFloat(timeSeries[date]["4. close"]));
      return { dates, prices };
    } catch (error) {
      console.error("Failed to load JSON file for " + symbol, error);
      return { dates: [], prices: [] };
    }
  }

  async function fetchTariffNews() {
    try {
      const res = await fetch("news.json");

      return [{
        date: new Date().toISOString().split("T")[0],
        title: "Unable to fetch news. Try again later.",
        description: ""
      }];
    }
  }

  function renderNews(allNews, dataArray) {
    const today = new Date();
    const threeDaysAgo = new Date(today);
    threeDaysAgo.setDate(today.getDate() - 3);

    let minDate = null, maxDate = null;
    dataArray.forEach(({ dates }) => {
      if (dates.length) {
        const dMin = new Date(dates[0]);
        const dMax = new Date(dates[dates.length - 1]);
        if (!minDate || dMin < minDate) minDate = dMin;
        if (!maxDate || dMax > maxDate) maxDate = dMax;
      }
    });

    const filteredNews = allNews
      .filter(article => {
        const newsDate = new Date(article.date);
        return newsDate >= minDate && newsDate <= maxDate;
      })
      .sort((a, b) => new Date(b.date) - new Date(a.date));

    const newsContainer = document.getElementById("newsList");
    newsContainer.innerHTML = "";

    if (filteredNews.length === 0) {
      newsContainer.innerHTML = "<p>No relevant news found in the last 30 days.</p>";
      return;
    }

    filteredNews.forEach(article => {
      const div = document.createElement("div");
      const newsDate = new Date(article.date);

      div.className = "news-item" + (isNew ? " new" : "");
      const dateStr = newsDate.toLocaleDateString('en-US', { year: 'numeric', month: 'short', day: 'numeric' });
